```python
files = ["MutualFundprices-A-E.csv",
         "MutualFundprices-F-K.csv",
         "MutualFundprices-L-P.csv",
         "MutualFundprices-Q-Z.csv"]

dfs = list()
for f in files:
    data = pd.read_csv(f)
    dfs.append(data)

df = pd.concat(dfs, ignore_index=True)
```

```python
df
```

```python
df['price_date'] = pd.to_datetime(df['price_date'])

# Pivot işlemi
pivot_df = df.pivot(index='price_date', columns='fund_symbol', values='nav_per_share')
df
```

```python
pivot_df
```

```python
mut = pd.read_csv("mut.csv")
mut.head()
```

```python
selected_symbols = mut['fund_symbol'].tolist()
pivot_df = pivot_df[selected_symbols]
columns_to_drop = [col for col in pivot_df.columns if col not in selected_symbols]
pivot_df.drop(columns=columns_to_drop, inplace=True)
```

```python
pivot_df
```

```python
pivot_df
```

```python
row_fill_percentage = pivot_df.count(axis=1) / len(pivot_df.columns) * 100

# %50'den fazla dolu olan ilk satırı bul
first_row_above_50_percent = pivot_df[row_fill_percentage > 50].iloc[0]

# Sonucu göster
print(first_row_above_50_percent)
```

```python
date_to_keep = '2010-12-27'
pivot_df = pivot_df.loc[date_to_keep:]
```
