```python
nanlist = mut_df.isna().sum()
print(nanlist["environment_score"])
print(nanlist["social_score"])
print(nanlist["governance_score"])
```

```python
mut_df.dropna(subset=["governance_score"], inplace=True)
```

```python
numeric_columns = mut_df.select_dtypes(include=np.number).columns

correlation_matrix = mut_df[numeric_columns].corr()

columns_to_drop = []
threshold = 0.4

for column in numeric_columns:
    if column not in ['initial_investment', 'environment_score', 'social_score', 'governance_score']:  # X, Y, Z sütunlarına kendi aralarındaki korelasyonları kontrol etmeyeceğiz
        correlation_with_i = correlation_matrix.loc[column, 'initial_investment']
        correlation_with_X = correlation_matrix.loc[column, 'environment_score']
        correlation_with_Y = correlation_matrix.loc[column, 'social_score']
        correlation_with_Z = correlation_matrix.loc[column, 'governance_score']
        
        if abs((correlation_with_i) < threshold or correlation_with_X) < threshold or abs(correlation_with_Y) < threshold or abs(correlation_with_Z) < threshold:
            columns_to_drop.append(column)

df = mut_df.drop(columns=columns_to_drop)
```

```python
import re
import ast
# Boş sütunları oluştur
for i in range(1, 11):
    mut_df[f'company_{i}'] = ''
    mut_df[f'ratio_{i}'] = 0.0

# Her bir satırdaki şirket ve oranları çıkaran yeni sütunlara verileri yerleştir
for i, row in mut_df.iterrows():
    companies = []
    ratios = []

    # Şirketleri (,) ile bölüp oranları ":" dan sonrası olarak al
    regex_pattern = re.compile(r'(?P<company>.*?):\s*(?P<ratio>\d+\.\d+)')

    # Regex ile eşleşenleri bul
    matches = regex_pattern.finditer(row['top10_holdings'])
    
    for j, match in enumerate(matches):
        company_name = match.group('company').strip()
        ratio = float(match.group('ratio'))

        # Verileri boş sütunlara yerleştir
        mut_df.at[i, f'company_{j + 1}'] = company_name
        mut_df.at[i, f'ratio_{j + 1}'] = ratio

```

```python
all_companies_dict = {}
for col in mut_df.columns:
    if 'company' in col:
        for company in mut_df[col]:
            if company:
                all_companies_dict[company] = len(all_companies_dict) + 1

# Her bir şirket sütununu ayrı ayrı ele alarak değerleri sözlükteki değerlerle değiştir
for col in mut_df.columns:
    if 'company' in col:
        mut_df[col] = mut_df[col].apply(lambda x: all_companies_dict[x] if x else x)
```
