```python
import yfinance as yf
import pandas as pd
```

```python
navsharedf = pd.read_csv("nav_per_share.csv")
```

```python
navsharedf
```

```python
threshold = 700
missing_values = navsharedf.isna().sum()
columns_to_drop = missing_values[missing_values > threshold].index
navsharedf = navsharedf.drop(columns=columns_to_drop)
navsharedf
```

```python
last_row = navsharedf.iloc[-1]
print(last_row.price_date)
```

```python
navsharedf = navsharedf.rename(columns={'price_date': 'Datetime'})
navsharedf.set_index("Datetime", inplace=True)
navsharedf
```

```python
fundList = navsharedf.columns.to_list()
```

```python
download_df = pd.DataFrame(columns=navsharedf.columns)
```

```python

def add_stock(index, main_df):
    new_df = yf.download(index, start="2021-12-01")
    new_df.rename_axis(index='Datetime')
    new_df = new_df.rename(columns={'Open': index})
    new_df = new_df[[index]].rename(columns={index: f'{index}_open'})
    main_df = pd.concat([main_df, new_df], axis=1, join='outer')
    
    return main_df
```

```python
for fund in fundList:
    download_df = add_stock(fund, download_df)
```

```python
download_df = download_df.dropna(axis=1, how='all')
download_df.columns = download_df.columns.str.rstrip('_open')
navsharedf.reset_index(inplace=True)
download_df.reset_index(inplace=True)
download_df = download_df.rename(columns={'index': 'Datetime'})
result_df = pd.concat([navsharedf, download_df], axis=0, ignore_index=True)
result_df['Datetime'] = pd.to_datetime(result_df['Datetime'])
result_df.set_index("Datetime", inplace=True)
```

```python
result_df.to_csv("current_nav.csv")
```
